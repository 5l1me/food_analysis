## Удаляем, то что не является странанми. 

```python
not_countries = [
    'Africa', 'Africa (FAO)',
    'Americas (FAO)', 'Asia',
    'Asia (FAO)', 'Belgium-Luxembourg (FAO)',
    'Caribbean (FAO)', 'Central America (FAO)',
    'Central Asia (FAO)', 'China (FAO)',
    'Eastern Africa (FAO)', 'Eastern Asia (FAO)',
    'Eastern Europe (FAO)', 'Europe',
    'Europe (FAO)', 'European Union (27)',
    'European Union (27) (FAO)', 'French Guiana',
    'High-income countries', 'Land Locked Developing Countries (FAO)',
    'Least Developed Countries (FAO)', 'Low Income Food Deficit Countries (FAO)',
    'Low-income countries', 'Lower-middle-income countries',
    'Melanesia', 'Micronesia (FAO)',
    'Middle Africa (FAO)', 'Net Food Importing Developing Countries (FAO)',
    'North America', 'Northern Africa (FAO)',
    'Northern America (FAO)', 'Northern Europe (FAO)',
    'Oceania', 'Oceania (FAO)',
    'Puerto Rico', 'Reunion',
    'Small Island Developing States (FAO)', 'South America',
    'South America (FAO)', 'South-eastern Asia (FAO)',
    'Southern Africa (FAO)', 'Southern Asia (FAO)',
    'Southern Europe (FAO)', 'Upper-middle-income countries',
    'Western Africa (FAO)','Western Asia (FAO)',
    'Western Europe (FAO)', 'World'
]

df = df[~df['Entity'].isin(not_countries)]
df.rename(columns={'Entity': 'Country'}, inplace=True)
```

## График всего продовольствия в мире

```python
# Список столбцов, которые представляют производство продовольствия
food_columns = ['Maize Production (tonnes)', 'Rice  Production ( tonnes)', 'Yams  Production (tonnes)', 
                'Wheat Production (tonnes)', 'Tomatoes Production (tonnes)', 'Tea  Production ( tonnes )', 
                'Sweet potatoes  Production (tonnes)', 'Sunflower seed  Production (tonnes)', 
                'Sugar cane Production (tonnes)', 'Soybeans  Production (tonnes)', 'Rye  Production (tonnes)', 
                'Potatoes  Production (tonnes)', 'Oranges  Production (tonnes)', 
                'Peas, dry Production ( tonnes)', 'Palm oil  Production (tonnes)', 
                'Grapes  Production (tonnes)', 'Coffee, green Production ( tonnes)', 
                'Cocoa beans Production (tonnes)', 'Meat, chicken  Production (tonnes)', 
                'Bananas  Production ( tonnes)', 'Avocados Production (tonnes)', 'Apples Production (tonnes)']

# Создание нового столбца 'Total Food Production', который является суммой всех столбцов продовольствия
df['Total Food Production'] = df[food_columns].sum(axis=1)

# Группировка данных по годам и подсчет общего производства продовольствия для каждого года
grouped = df.groupby('Year')['Total Food Production'].sum().reset_index()

# Создание графика
plt.figure(figsize=(10, 6))
plt.plot(grouped['Year'], grouped['Total Food Production'])
plt.title('График производства всего продовольствия')
plt.xlabel('Год')
plt.ylabel('Продовольствие (тонны)')
plt.show()


```

## Теперь рассмотрим производство по годам, каждой конкретной продукции

```python
plt.figure(figsize=(12, 6))  # устанавливаем размер графика
for col in food_columns:
    temp = df.groupby('Year')[col].sum()
    plt.plot(temp.index, temp.values, label=col)   
plt.xlabel('Year')
plt.ylabel('Общее производство продовольствия (тонны)')
plt.title('Общее производство продовольствия')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), prop={'size': 10})  # изменяем размер и положение легенды
plt.grid()
plt.tight_layout()
plt.show()
```

## Для примера рассмотри рост производства пшеницы в 1990-1992 по странам

```python
for year in [1990, 1991, 1992]:
    df_year = df[df['Year'] == year]
    df_year = df_year[df_year['Wheat Production (tonnes)'] > 10**7]  # фильтруем данные для наглядности

    plt.figure(figsize=(10, 6))
    plt.bar(df_year['Country'], df_year['Wheat Production (tonnes)'])
    plt.title(f'Производство пшеницы в {year}')
    plt.xlabel('Страна')
    plt.ylabel('Производство пшеницы (тонны)')
    plt.xticks(rotation=90)
    plt.show()
```

 _Видим, что островное государство Сент-Люсия произвела черезвучайно много пшеницы в 1991 и 1992 году, хотя их экономика базируется на туризме. Видимо это аномалия._

```python
# Выберем данные для 2012 года
df_2012 = df[df['Year'] == 2012]

# Создаем график
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_2012[df.columns[7]], df_2012[df.columns[18]], c=df_2012.index, cmap='viridis')
plt.title("Производство чая и кофе в разных странах в 2012 году")
plt.xlabel("Производство чая (тонны)")
plt.ylabel("Производство кофе (тонны)")

# Добавляем легенду
legend1 = plt.legend(*scatter.legend_elements(), title="Страна")
plt.gca().add_artist(legend1)

plt.show()
```

_Опять видны черезвучайно большие числа в производстве чая у Чада и Беларуси(точки в правом нижнем угле), при этом Китая, который приянято считать лидером, на графике не заметно._

## Сравним производство всей продукции СССР в 1970 и постсоветских стран в 2015

```python
post_soviet_countries = ['Armenia', 'Azerbaijan', 'Belarus', 'Estonia', 'Georgia',
                         'Kazakhstan', 'Kyrgyzstan', 'Latvia', 'Lithuania', 'Moldova',
                         'Russia', 'Tajikistan', 'Turkmenistan', 'Ukraine', 'Uzbekistan']

# Выберем данные для СССР в 1970 году и постсоветских стран в 2015 году
df_1970 = df[(df['Country'] == 'USSR') & (df['Year'] == 1970)]
df_2015 = df[(df['Country'].isin(post_soviet_countries)) & (df['Year'] == 2015)]

# Объединим эти два набора данных
df_compare = pd.concat([df_1970, df_2015])

# Подсчитаем общее производство продовольствия для каждой страны
df_compare['Total Food Production'] = df_compare[food_columns].sum(axis=1)

# Создаем график
plt.figure(figsize=(10, 6))
for year in df_compare['Year'].unique():
    df_year = df_compare[df_compare['Year'] == year]
    plt.bar(df_year['Country'], df_year['Total Food Production'], label=str(year))

plt.title("Сравнение общего производства продовольствия в СССР в 1970 году и постсоветских странах в 2015 году")
plt.xlabel("Страна")
plt.ylabel("Производство продовольствия (тонны)")
plt.legend(title="Год")
plt.xticks(rotation=90)
plt.show()
```

_Сразу виднеется огромная аномалия у Армении_
